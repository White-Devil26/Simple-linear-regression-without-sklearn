### Spliting dataset into dependent & independent variables

```python
X = dataset.iloc[:, 0].values 
Y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 1/3, random_state = 0)
```

```python
theta0 = 0 #parameter
theta1 = 0 #parameter
aplpha = 0.05 #learning rate
L = len(X_train) #length of the training set
for i in range(L):
    Y_prediction = theta0 + theta1 * X_train #linear equation
    # cost function
    J = (1/2*L)*sum(Y_prediction - Y_train)
    
    # gradient decent 
    Dev_theta0 = (-1/L) * sum(Y_prediction - Y_train)
    Dev_theta1 = (-1/L) * sum( X_train * (Y_prediction - Y_train))
    
    #updating parameter
    
    theta0 = theta0 + aplpha * Dev_theta0
    theta1 = theta1 + aplpha * Dev_theta1
    
print(theta0,'\n',theta1)
Y_prediction = theta0 + theta1 * X_train
```

```python
plt.scatter(X_train,Y_train)
plt.plot([min(X_train),max(X_train)], [min(Y_prediction),max(Y_prediction)])
```

```python
plt.scatter(X_test,Y_test)
plt.plot([min(X_test),max(X_test)], [min(Y_prediction),max(Y_prediction)])
```
